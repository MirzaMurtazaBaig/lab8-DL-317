```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r'C:\Users\SED\Downloads\haider-zalghout-bvqbuG8vMtE-unsplash.jpg')  
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  

brightness_increase = 30  # Value from 0 to 255
bright_image = cv2.convertScaleAbs(image, alpha=1, beta=brightness_increase)
bright_image_rgb = cv2.cvtColor(bright_image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(bright_image_rgb)
plt.title("Brightened Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
